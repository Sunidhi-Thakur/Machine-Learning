# Missing Values Imputation 
#### Measure of central tendency (Mode)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("hp_train.csv")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

df_cat = df.select_dtypes(include='object')
df_cat.shape
```

```python
missing_val_per = df_cat.isnull().sum()/df_cat.shape[0] * 100
missing_val_per
```

```python
drop_vars = ['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
df2_cat = df_cat.drop(columns = drop_vars, axis = 1)
df2_cat.shape
```

```python
isnull_per = df2_cat.isnull().mean() * 100
missing_vars = isnull_per[isnull_per > 0].keys()
missing_vars
```

```python
df2_cat['MasVnrType'].value_counts()
```

```python
df2_cat['MasVnrType'].mode()
```

```python
df2_cat['MasVnrType'].fillna(df2_cat['MasVnrType'].mode()[0])
```

```python
for var in missing_vars:
    df2_cat[var].fillna(df2_cat[var].mode()[0], inplace = True)
    print(var,": ",df2_cat[var].mode()[0])
```

```python
df2_cat.isnull().sum().sum()
```

```python
plt.figure(figsize = (16,9))
for i,var in enumerate(missing_vars):
    plt.subplot(4,3,i+1)
    plt.hist(df2_cat[var], label = 'Cleaned')
    plt.hist(df[var].dropna(), label = 'Original')
    plt.legend()
```

```python
df.update(df2_cat)
df.drop(columns = drop_vars, inplace = True)
```
