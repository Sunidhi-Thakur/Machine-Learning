# Data Cleaning
* Fill missing value manually
* Global constant
* Measure of central tendency (Mean and Median)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset_path = "hp_train.csv"
df = pd.read_csv(dataset_path)
```

```python
df.info()
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
df.isnull().sum()
```

```python
null_var = df.isnull().sum()/df.shape[0] * 100
null_var
```

```python
drop_columns = null_var[null_var > 20].keys()
drop_columns
```

```python
df_drop_col = df.drop(columns = drop_columns)
df_drop_col.shape
```

```python
df_num = df_drop_col.select_dtypes(include=['int64','float64'])
df_num.head()
```

```python
plt.figure(figsize = (25,25))
sns.heatmap(df_num.isnull())
```

```python
df_num[df_num.isnull().any(axis = 1)]
```

```python
df_num.isnull().sum()
```

```python
missing_num_var = [var for var in df_num.columns if df_num[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize = (10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.histplot(df_num[var], bins = 20, kde = True)
```

```python
df_mean = df_num.fillna(df_num.mean())
df_mean.isnull().sum().sum()
```

```python
plt.figure(figsize = (10,10))
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.histplot(df_num[var], bins = 20, kde = True, color = 'b', label = 'Original')
    sns.histplot(df_mean[var], bins = 20, kde = True,color = 'r', label = 'Mean' )
    plt.legend()
```

```python
df_median = df_num.fillna(df_num.median())
df_median.isnull().sum().sum()
```

```python
plt.figure(figsize = (10,10))
sns.set_style("whitegrid", {'axes.grid' : False})

for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.histplot(df_num[var], bins = 20, kde = True, color = 'b', label = 'Original')
    sns.histplot(df_mean[var], bins = 20, kde = True,color = 'g', label = 'Mean' )
    sns.histplot(df_median[var], bins = 20, kde = True,color = 'r', label = 'Median' )
    plt.legend()
```

```python
for i,var in enumerate(missing_num_var):
    plt.figure(figsize = (10,10))
    plt.subplot(311)
    sns.boxplot(x=df[var])
    plt.subplot(312)
    sns.boxplot(x=df_mean[var], color = 'r')
    plt.subplot(313)
    sns.boxplot(x=df_median[var], color = 'g')
```

```python
df_concat = pd.concat([df_num[missing_num_var],df_mean[missing_num_var], df_median[missing_num_var]], keys = ['Original', 'Mean','Median'], axis = 1)
```
