### Dummy Variables and One-Hot Encoding

```python
import pandas as pd
import seaborn as sns
```

```python
tips_df = sns.load_dataset("tips")
tips_df.tail()
```

```python
tips_df.shape
```

```python
dummy_df = pd.get_dummies(tips_df)
dummy_df
```

```python
dummy_df2 = pd.get_dummies(tips_df, drop_first=True)
dummy_df2
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc = OneHotEncoder(sparse = False)
```

```python
enc_arr = enc.fit_transform(tips_df[['sex','smoker','day','time']])
enc_arr
```

```python
dummy_df.keys()
```

```python
enc_df = pd.DataFrame(enc_arr, columns = ['sex_Male', 'sex_Female', 'smoker_Yes',
       'smoker_No', 'day_Thur', 'day_Fri', 'day_Sat', 'day_Sun', 'time_Lunch',
       'time_Dinner'])
enc_df
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("hp_train.csv")
df.head()
```

```python
df2 = df[['KitchenQual','BldgType']]
```

```python
enc = LabelEncoder()
```

```python
enc.fit_transform(df['BldgType'])
```

```python
df2['BldgType_enc'] = enc.fit_transform(df['BldgType'])
df2
```

```python
df2['BldgType_enc'].value_counts()
```

```python
df2['BldgType'].value_counts()
```

```python
df2['KitchenQual'].value_counts()
```

```python
order_label = {'Ex':3, 'Gd':2, 'TA':1, 'Fa':0}
```

```python
df2['KitchenQual_enc'] = df2['KitchenQual'].map(order_label)
df2
```

```python
df2['KitchenQual_enc'].value_counts()
```
