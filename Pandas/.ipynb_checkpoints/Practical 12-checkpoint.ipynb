### Pandas Merge

```python
import pandas as pd
df1 = pd.DataFrame({'Roll no':[1,2,3,4],
                    'Class':[9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'Roll no':[1,2,3,4],
                    'Name':['Caitlin','Barry','Iris','Cisco']})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on = 'Roll no')
```

```python
pd.merge(df2, df1, on = 'Roll no')
```

### Different Roll no in Dataframe

```python
df1 = pd.DataFrame({'Roll no':[1,2,3,4],
                    'Class':[9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'Roll no':[1,2,3,5],
                    'Name':['Caitlin','Barry','Iris','Cisco']})
df2
```

```python
pd.merge(df2, df1, on = 'Roll no')
```

```python
pd.merge(df1, df2, how = 'outer')
```

```python
pd.merge(df1, df2, how = 'left')
```

```python
pd.merge(df1, df2, how = 'right')
```

```python
pd.merge(df1, df2, how = 'inner')
```

```python
  pd.merge(df1, df2, how = 'outer', indicator = True)
```

```python
df1 = pd.DataFrame({'Roll no':[1,2,3,4],
                    'Class':[9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'Roll no':[1,2,3,4],
                    'Class':[9,10,11,12]})
df2
```

```python
pd.merge(df1, df2, on = 'Roll no')
```

```python
pd.merge(df1, df2, on = 'Roll no', suffixes = ('-Higher', '_Middle'))
```
