## Concat()

### Series

```python
import pandas as pd
```

```python
s1 = pd.Series([0,1,2])
s1
```

```python
s2 = pd.Series([3,4,5])
s2
```

```python
pd.concat([s1, s2])
```

#### Unequal Series

```python
s2 = pd.Series([3,4,5,6,7,8])
s2
```

```python
pd.concat([s1, s2])
```

### DataFrame

```python
df1 = pd.DataFrame({'ID': [1,2,3], 
                   'Name': ['A','B','C'],
                   'Class': ['C1','C2','C3']})
df1
```

```python
df2 = pd.DataFrame({'ID': [4,5,6,7], 
                   'Name': ['D','E','F','G'],
                   'Class': ['C1','C2','C3', 'C4']})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df2, df1])   # axis = 0 [row]
```

```python
pd.concat([df1, df2], axis = 1)  # axis = 1 [column]
```

```python
pd.concat([df1, df2], ignore_index = True )
```

```python
pd.concat([df1, df2])
```

```python
df1 = pd.DataFrame({'ID': [1,2,3], 
                   'Name': ['A','B','C'],
                   'Class': ['C1','C2','C3']})
df1
```

```python
df2 = pd.DataFrame({'ID': [2,1], 
                   'Name': ['B','C'],
                   'Class': ['C1','C3']})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 1)
```

```python
pd.concat([df1, df2], axis = 1, join = 'inner')
```

```python
df1 = pd.DataFrame({'ID': [1,2,3], 
                   'Name': ['A','B','C'],
                   'Class': ['C1','C2','C3']})
df1
```

```python
df2 = pd.DataFrame({'ID': [4,5,6], 
                   'Name': ['D','E','F'],
                   'Class': ['C1','C2','C3']})
df2
```

```python
pd.concat([df1, df2], keys = ['First', 'Second'])
```

```python
pd.concat([df1, df2], keys = ['First', 'Second'], axis = 1)
```

```python
df1 = pd.DataFrame({'ID': [1,2,3], 
                   'Name': ['A','B','C'],
                   'Class': ['C1','C2','C3']})
df1
```

```python
df2 = pd.DataFrame({'Marks': [98, 93, 89]})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], sort = False)
```

```python
pd.concat([df1, df2], sort = True)
```
