## Join()

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4],
                   'B':[10,20,30,40]})
df2 = pd.DataFrame({'C': [5,6,7],
                   'D':[50,60,70]})
```

```python
display(df1, df2)
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4],
                   'B':[10,20,30,40]}, index = ['a','b','c','d'])
df2 = pd.DataFrame({'C': [5,6,7],
                   'D':[50,60,70]})
display(df1, df2)
```

```python
df1.join(df2)
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4],
                   'B':[10,20,30,40]}, index = ['a','b','c','d'])
df2 = pd.DataFrame({'C': [5,6,7],
                   'D':[50,60,70]}, index = ['a','b','c'])
display(df1, df2)
```

```python
df1.join(df2)
```

```python
df1.join(df2, how = 'right')
```

```python
df1.join(df2, how = 'inner')
```

```python
df1.join(df2, how = 'outer')
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4],
                   'B':[10,20,30,40]})
df2 = pd.DataFrame({'A': [5,6,7],
                   'C':[50,60,70]})
display(df1, df2)
```

```python
df1.join(df2, lsuffix = '_1')
```

```python
df1.join(df2, rsuffix = '_1')
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4],
                   'B':[10,20,30,40]})
df2 = pd.DataFrame({'A': [5,6,7],
                   'B':[50,60,70]})
display(df1, df2)
```

```python
df1.join(df2, rsuffix = '_1')
```
