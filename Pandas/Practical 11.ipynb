### Pandas GroupBy

```python
import pandas as pd
df = pd.read_csv('C:\\Users\\sunid\\Machine Learning\\Pandas\\student_result2.csv')
df
```

```python
grp1 = df.groupby(by = 'Section')
```

```python
grp1.groups
```

```python
df.groupby(['Section', 'Class']).groups
```

```python
for Section, df_1 in grp1:
    print(Section)
    print(df_1)
    print('\n')
```
