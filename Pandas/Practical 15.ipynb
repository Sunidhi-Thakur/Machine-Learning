## Append()

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': [10,20,30],
                   'B': [1,2,3]})
df2 = pd.DataFrame({'A': [50,70,60],
                   'B': [4,5,6]})
display(df1, df2)
```

```python
df1.append(df2)
```

```python
df1.append(df2, ignore_index = True)
```
