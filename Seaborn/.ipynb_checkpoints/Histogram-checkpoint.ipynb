# Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips_df = sns.load_dataset("tips")
tips_df.tail()
```

```python
sns.displot(tips_df["size"])
```

```python
sns.displot(tips_df["total_bill"])
```

```python
sns.displot(tips_df["total_bill"], bins = 100)
```

```python
sns.displot(tips_df["total_bill"], kind = 'kde')  #kernel density estimator
```

```python
sns.displot(tips_df["total_bill"], kind = 'kde', rug = True)
```

```python
sns.displot(tips_df["total_bill"], color = 'r')
```

```python
sns.set()
sns.displot(tips_df["total_bill"], color = 'g')
```

```python
tips_df.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]

plt.figure(figsize = (16,9))
sns.set() 

sns.displot(tips_df["total_bill"], bins = bins)

plt.xticks(bins)

plt.title("Histogram", fontsize = 25)
plt.show()
```
