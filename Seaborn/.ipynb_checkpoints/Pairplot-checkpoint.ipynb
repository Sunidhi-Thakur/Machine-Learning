## Pairplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                        columns = np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean perimeter', 'mean smoothness', 'mean area', 'mean texture'])
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean perimeter', 'mean smoothness', 'mean area', 'mean texture'],
            hue = 'target')
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean perimeter', 'mean smoothness', 'mean area', 'mean texture'],
            hue = 'target', hue_order = [1.0, 0.0])
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean perimeter', 'mean smoothness', 'mean area', 'mean texture'],
            hue = 'target', palette = 'inferno')
```

```python
sns.pairplot(cancer_df, hue = 'target', x_vars = ['mean radius', 'mean perimeter'], y_vars = ['mean radius'])
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean perimeter', 'mean smoothness', 'mean area', 'mean texture'],
            hue = 'target', kind = 'reg')
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean perimeter', 'mean smoothness', 'mean area', 'mean texture'],
            hue = 'target', diag_kind = 'hist')
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean perimeter', 'mean smoothness', 'mean area', 'mean texture'],
            hue = 'target', markers = ['*','^'])
```
