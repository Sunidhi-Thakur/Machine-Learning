# Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
titanic_df = sns.load_dataset("titanic")
titanic_df.head()
```

```python
titanic_df.shape
```

```python
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df)
plt.show()
```

```python
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df)
plt.show()
```

```python
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df,
               style = 'who')
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.scatterplot(x = 'age', y = 'fare', hue = 'sex', data = titanic_df,
               style = 'who', size = 'who', sizes = (90,200))
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.scatterplot(x = 'who', y = 'fare', hue = 'alive', data = titanic_df,
               style = 'alive', size = 'who', sizes = (90,200))
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.scatterplot(x = 'who', y = 'fare', hue = 'alive', data = titanic_df,
               style = 'alive', size = 'who', sizes = (100,400), palette = 'pastel')
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df)
sns.lineplot(x = 'age', y = 'fare', data = titanic_df)
sns.barplot(x = 'age', y = 'fare', data = titanic_df)
```
