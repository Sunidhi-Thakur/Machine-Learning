## Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
tips_df = sns.load_dataset("tips")
tips_df.tail()
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill)
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill, hue = tips_df.sex)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df)
```

```python
order = ['Sun', 'Sat', 'Thur', 'Fri']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, order = order)
```

```python
hue_order = ['Female', 'Male']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, hue_order = hue_order)
```

```python
hue_order = ['Female', 'Male']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, hue_order = hue_order, palette = 'inferno')
```

```python
hue_order = ['Female', 'Male']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, hue_order = hue_order, estimator = np.max, palette = 'winter')
```

```python
hue_order = ['Female', 'Male']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, hue_order = hue_order, estimator = np.max, palette = 'Accent', ci = 50)
```

```python
sns.barplot(x = 'total_bill', y = 'day', hue = 'sex', data = tips_df)
```

```python
sns.barplot(x = 'total_bill', y = 'size', hue = 'sex', data = tips_df,
           orient = 'h')
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, color = 'r')
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, saturation = 0.6)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, errcolor = '0.5')
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, errwidth = 12)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, capsize = 1)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, dodge = False)
```

```python
sns.set()
kwargs = {'alpha':0.9, 'linestyle': '--', 'linewidth': 2, 'edgecolor': 'k'}

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, **kwargs)
```

```python
sns.set()
kwargs = {'alpha':0.9, 'linestyle': '--', 'linewidth': 2, 'edgecolor': 'k'}

sns.barplot(x = 'day', y = 'total_bill', data = tips_df, **kwargs)
```

```python
sns.barplot(x = 'day', y='total_bill',
           data = tips_df, alpha = 0.4, linestyle = ':', linewidth = 6,
           edgecolor = 'k')
```

```python
ax = sns.barplot(x = 'day', y='total_bill',
           data = tips_df, alpha = 0.4, linestyle = ':', linewidth = 6,
           edgecolor = 'k')
ax.set(title = "Bar Plot",
      xlabel = "day",
      ylabel = 'total_bill')
```

```python
plt.figure(figsize = (16,9))

sns.barplot(x = 'day', y='total_bill',
           data = tips_df, linestyle = '--', linewidth = 2,
           edgecolor = 'k')
plt.title("Bar Plot", fontsize = 25)
plt.xlabel("day", fontsize = 15)
plt.ylabel('total_bill', fontsize = 15)
```
