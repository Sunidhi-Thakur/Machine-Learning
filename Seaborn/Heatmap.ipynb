# Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)
```

```python
global_df = pd.read_csv('global_warming.csv')
global_df.head()
```

```python
global_df.shape
```

```python
global_df = global_df.drop(columns = ["Country Code", "Indicator Name", "Indicator Code"], axis = 1).set_index("Country Name")
global_df.head()
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(global_df)
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(global_df, vmin = 0, vmax = 21)
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(global_df, cmap = 'icefire')
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(global_df, annot = True, cmap = 'inferno')
```

```python
annot_arr = np.array([['a00','a01','a02'], ['a10','a11','a12'], ['a20','a21','a22'], ['a30','a31','a32']]).reshape(4,3)
```

```python
sns.heatmap(arr_2d, annot = annot_arr, fmt = 's')
```

```python
plt.figure(figsize = (16,9))

annot_kws = {'fontsize': 10,
            'fontstyle': 'italic',
            'color':'k',
            'alpha': 0.8,
            'rotation': 'vertical',
            'verticalalignment': 'center',
            'backgroundcolor': 'w'}
sns.heatmap(global_df, cmap = 'coolwarm', annot = True, annot_kws = annot_kws)
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(global_df, annot = True, linewidth = 4)
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(global_df, annot = True, cbar = False)
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(global_df, annot = True, cbar = False, xticklabels = False, yticklabels = False)
```

```python
plt.figure(figsize = (16,9))

cbar_kws = {'orientation': 'horizontal',
            'shrink':1,
            'extend': 'min',
            'extendfrac': 0.5,
            'ticks': np.arange(0,22),
            'drawedges': True
           }
sns.heatmap(global_df, cbar_kws = cbar_kws)
```

```python
plt.figure(figsize = (16,9))

country_code = ['USA','GBR','IND','CHN', 'RUS','AUS','FRA','DEU','CAN','BRA','ARG','PAK','NPL','BGD','JPN']
sns.heatmap(global_df, xticklabels = np.arange(0,15), yticklabels = country_code)
```

```python
plt.figure(figsize = (16,9))

ax = sns.heatmap(global_df, linewidth = 0.5, cmap = 'coolwarm')

ax.set(title= 'Who is responsible for Global Warming?',
      xlabel='CO2 emissions (metric tons per capita)',
      ylabel='Country Name')
sns.set(font_scale = 5)
```

## Correlation

A *correlation* is a statistical measure of the relationshi between two variables (X,Y).<br>
The correlation coefficient is value from -1 to 1.<br>
&emsp;&emsp;1. -1: Perfect negative correlation (X increases, Y decreases)<br>
&emsp;&emsp;2. 0: No correlation<br>
&emsp;&emsp;3. 1: Perfect positive correlation (X increases, Y increases)<br>
    <hr>
    

    

```python
corr = global_df.corr()
corr
```

```python
plt.figure(figsize=(16,9))

sns.set(font_scale = 1)

ax = sns.heatmap(global_df.corr(), annot = True, linewidth = 3)
ax.tick_params(size = 10, color = 'w', labelsize = 10, labelcolor = 'b')
plt.title('Who is responsible for Global Warming?', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(16,9))         
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights, annot = True)
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                        columns = np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize = (30,30))

ax = sns.heatmap(cancer_df.corr(), annot = True, linewidth = 3)
ax.tick_params(size = 10, color = 'w', labelsize = 10, labelcolor = 'k')
plt.title('Breast Cancer Detection', fontsize = 20)
plt.show()
```
