# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
classes = ['Python', 'ML', 'Java', 'AI', 'C++']
n1 = [20, 27, 30, 25, 4]
```

```python
plt.pie([1])
```

```python
classes = ['Python', 'ML', 'Java', 'AI', 'C++']
n1 = [20, 27, 30, 25, 14]
explode = [0,0.05,0,0,0]
plt.pie(n1, labels = classes, explode = explode)
plt.show()
```

```python
colors = ['skyblue','pink','slateblue','lightgreen','brown']
explode = [0,0.05,0,0,0]
plt.pie(n1, labels = classes, explode = explode, colors = colors)
plt.show()
```

```python
colors = ['skyblue','pink','slateblue','lightgreen','brown']
explode = [0,0.05,0,0,0]
plt.pie(n1, labels = classes, explode = explode, colors = colors, autopct = '%0.2f%%')
plt.show()
```

```python
explode = [0,0.05,0,0,0]
plt.pie(n1, labels = classes, explode = explode, shadow = True)
plt.show()
```

```python
colors = ['skyblue','pink','slateblue','lightgreen','brown']
explode = [0,0.05,0,0,0]
plt.pie(n1, labels = classes, explode = explode, colors = colors, autopct = '%0.2f%%', radius = 2)
plt.show()
```

```python
colors = ['skyblue','pink','slateblue','lightgreen','brown']
explode = [0,0.05,0,0,0]
plt.pie(n1, labels = classes, explode = explode, colors = colors, autopct = '%0.2f%%', startangle = 270)
plt.show()
```

```python
explode = [0,0.05,0,0,0]
textprops = {'fontsize':12}
plt.pie(n1, labels = classes, explode = explode, textprops = textprops)
plt.show()
```

```python
wedgeprops = {"linewidth":3, "width":1, "edgecolor": 'k'}
plt.pie(n1,
       labels = classes,
       explode = explode,
       autopct = '%0.2f%%',
       pctdistance = 0.6,
       shadow = True,
       labeldistance = 1.2,
       wedgeprops = wedgeprops,
       radius = 1.5)
```

```python
wedgeprops = {"linewidth":3, "width":4, "edgecolor": 'k'}
plt.pie(n1,
       labels = classes,
       explode = explode,
       autopct = '%0.2f%%',
       pctdistance = 0.6,
       shadow = True,
       labeldistance = 1.2,
       wedgeprops = wedgeprops,
       radius = 1.5)
```

```python
plt.figure(figsize = (6,4))
wedgeprops = {"linewidth":3, "width":1, "edgecolor": 'k'}
plt.pie(n1,
       labels = classes,
       explode = explode,
       autopct = '%0.2f%%',
       pctdistance = 0.6,
       shadow = True,
       labeldistance = 1.2,
       radius = 1.5)
plt.legend()
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(5,3))
colors = ['r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w']
labels = np.ones(20)

plt.pie([1], colors = 'k', radius = 2.05)
plt.pie(labels, colors = colors, radius = 2.0)

plt.pie([1], colors = ['blue'], radius = 1.8)
plt.pie([1], colors = ['hotpink'], radius = 1.6)
plt.pie([1], colors = ['green'], radius = 1.3)
plt.pie([1], colors = ['maroon'], radius = 1.1)
plt.pie([1], colors = ['yellow'], radius = 0.9)

plt.pie([1], colors = 'b', radius = 0.31)
plt.pie(labels, colors = colors, radius = 0.3)

plt.pie([1], colors = 'w', radius = 0.2)
plt.pie([1], colors = 'k', radius = 0.1)

plt.show()
```
