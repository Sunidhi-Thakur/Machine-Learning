## Subplot

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.pie([1])

plt.subplot(2,2,2)
plt.pie([1,2])

plt.subplot(2,2,3)
plt.pie([1,3,4])

plt.subplot(2,2,4)
plt.pie([1,3,5,7])

plt.show()
```
