## Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
s1 = np.random.randint(20,60, (100))
s2 = np.random.randint(25, 35, (100))
print(s1)
print(s2)
```

```python
plt.hist(s1)
plt.title("Employees Age")
plt.ylabel('# of employees')
plt.xlabel('age')
```

```python
bins = [5,10,15,20,25,30,35,40,45,50,55,60,65]
plt.hist(s1, bins)
plt.title("Employees Age")
plt.ylabel('# of employees')
plt.xlabel('age')
```

```python
bins = [5,10,15,20,25,30,35,40,45,50,55,60,65]
plt.hist(s1, bins, rwidth = 0.8)
plt.title("Employees Age")
plt.ylabel('# of employees')
plt.xlabel('age')
```

```python
bins = [5,10,15,20,25,30,35,40,45,50,55,60,65]
plt.hist(s1, bins, histtype = "step")
plt.title("Employees Age")
plt.ylabel('# of employees')
plt.xlabel('age')
```

```python
plt.hist(s1, orientation = "horizontal", rwidth = 0.8)
plt.title("Employees Age")
plt.xlabel('# of employees')
plt.ylabel('age')
```

```python
plt.hist(s1, color = 'r')
plt.title("Employees Age")
plt.ylabel('# of employees')
plt.xlabel('age')
```

```python
plt.hist(s2, color = 'm', label = 'student')
plt.title("Students Age")
plt.ylabel('# of students')
plt.xlabel('age')
plt.legend()
```

```python
plt.hist(s2, color = 'b', label = 'student')
plt.hist(s1, color = 'm', label = 'employee', orientation = 'horizontal')
plt.title("Age Distribution")
plt.ylabel('#')
plt.xlabel('age')
plt.legend()
```

```python
plt.hist([s1,s2], color = ['b','g'], label = ['employee','student'])
plt.title("Age Distribution")
plt.ylabel('#')
plt.xlabel('age')
plt.legend()
```

```python
plt.figure(figsize = (16,9))
plt.hist([s1,s2], color = ['b','g'], label = ['employee','student'])
plt.title("Age Distribution")
plt.ylabel('#')
plt.xlabel('age')
plt.legend()
```

```python
from matplotlib import style
style.use('ggplot')

plt.figure(figsize = (16,9))
plt.hist([s1,s2], color = ['b','g'], label = ['employee','student'])
plt.title("Age Distribution")
plt.ylabel('#')
plt.xlabel('age')
plt.legend()
```
