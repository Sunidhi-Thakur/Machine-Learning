```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ['Python', 'ML', 'Java', 'AI', 'C++']
n1 = [20, 27, 30, 25, 4]  # no of students in each class
n2 = [15, 23, 45, 12, 26]
n3 = [30, 2, 12, 33, 3]
```

```python
plt.bar(classes, n1)
```

```python
plt.barh(classes, n1)
```

```python
plt.bar(classes, n1, width = 0.3)
```

```python
plt.bar(classes, n1, width = 0.3, align = 'edge')
```

```python
plt.bar(classes, n1, color = 'orange')
```

```python
plt.bar(classes, n1, edgecolor = 'red')
```

```python
plt.bar(classes, n1, edgecolor = 'red', linewidth = 3)
```

```python
plt.bar(classes, n1, edgecolor = 'red', linewidth = 3, alpha = 0.5)
```

```python
plt.bar(classes, n1, edgecolor = 'red', linewidth = 3, linestyle = '--')
```

```python
plt.bar(classes, n1, edgecolor = 'red', linewidth = 3, label = 'students in class 1')
plt.legend()
```

```python
style.use('ggplot')
plt.figure(figsize = (16,9))
plt.bar(classes, n1, label = 'students in class 1')
plt.legend()
plt.title("Classes", fontsize = 20)
plt.ylabel('# of students', fontsize = 15)
plt.xlabel('subjects', fontsize = 15)
```

```python
plt.figure(figsize = (16,9))
plt.bar(classes, n1, label = 'students in class 1', color = 'b')
plt.bar(classes, n2, label = 'students in class 2', color = 'm')
plt.bar(classes, n3, label = 'students in class 3', color = 'g')
plt.legend()
plt.title("Classes", fontsize = 20)
plt.ylabel('# of students', fontsize = 15)
plt.xlabel('subjects', fontsize = 15)
```

```python
plt.figure(figsize = (16,9))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index+width+width, n3, width, label = 'students in class 1', color = 'r', alpha = 0.65)
plt.bar(classes_index, n2, width, label = 'students in class 2', color = 'b', alpha = 0.65)
plt.bar(classes_index+width, n3, width, label = 'students in class 3', color = 'g', alpha = 0.65)
plt.legend()
plt.xticks(classes_index+width, classes, rotation = 30)
plt.title("Classes", fontsize = 20)
plt.ylabel('# of students', fontsize = 15)
plt.xlabel('subjects', fontsize = 15)
```

```python
classes_index = np.arange(len(classes))
width = 0.2
plt.barh(classes_index+width+width, n3, width, label = 'students in class 1', color = 'r', alpha = 0.65)
plt.barh(classes_index, n2, width, label = 'students in class 2', color = 'b', alpha = 0.65)
plt.barh(classes_index+width, n3, width, label = 'students in class 3', color = 'g', alpha = 0.65)
plt.legend()
plt.yticks(classes_index+width, classes, rotation = 30)
plt.title("Classes", fontsize = 20)
plt.xlabel('# of students', fontsize = 15)
plt.ylabel('subjects', fontsize = 15)
```
